```python
import pandas as pd
data = pd.read_csv("../data/PV_panel_power_with_weather_hour.csv", delimiter=";")
data = data.drop(["MODELICA", 'MODELICA_ERROR', 'SHADOW'], axis=1)
# df = data.copy(deep=True)
data['DATE'] = pd.to_datetime(data['DATE'])
```

```python
data.describe()
```

```python
# Extracting seasonal information
data.set_index('DATE', inplace=True)
data['MONTH'] = data.index.month
data['DAY_OF_YEAR'] = data.index.dayofyear
data['WEEK_OF_YEAR'] = data.index.isocalendar().week
data['DAY_OF_WEEK'] = data.index.dayofweek
data['HOUR'] = data.index.hour

# Selecting relevant columns for modeling
selected_features = ['IRRADIANCE_SOLAR', 'AIR_TEMPERATURE_CELSIUS', 'DIRECTION_WIND', 'IGRO', 'VELOCITY_WIND', 'RAFF_WIND', 'PV_Power',
            'MONTH', 'DAY_OF_YEAR', 'WEEK_OF_YEAR', 'DAY_OF_WEEK', 'HOUR']

data = data[selected_features].dropna()

# Smoothing all data
data = data.rolling(window=32, min_periods=1).mean()

# Scaling the data using MaxAbsScaler
scaler = MaxAbsScaler()
scaled_series = scaler.fit_transform(data)
```

```python
def create_ml_dataset(data, time_step=12, pv_power_index=6):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), :-1].flatten())  # Flatten features for ML models
        y.append(data[i + time_step, pv_power_index])  # Use the dynamically found index for 'PV_Power'
    return np.array(X), np.array(y)

time_step = 12
X, y = create_ml_dataset(scaled_series, time_step, pv_power_index = data.columns.get_loc('PV_Power'))

# TimeSeries cross-validator (k-fold)
tscv = TimeSeriesSplit(n_splits=5)
```

```python
# River Model - SNARIMAX (TimeSeries Model)
mse_list_river, rmse_list_river, mae_list_river = [], [], []
all_y_test_river, all_y_pred_river = [], []

# Iterate over folds for River model
for train_index, test_index in tscv.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Reinitialize the River SNARIMAX model for each fold
    river_model = SNARIMAX(p=1, d=0, q=1, m=12, sp=1, sd=0, sq=1)

    # Online training with River
    for yi in y_train:
        river_model.learn_one(yi)

    # Predictions
    y_pred = [river_model.forecast(horizon=1)[0] for _ in range(len(y_test))]

    # Metrics calculation
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)

    mse_list_river.append(mse)
    rmse_list_river.append(rmse)
    mae_list_river.append(mae)

    all_y_test_river.extend(y_test)
    all_y_pred_river.extend(y_pred)

# Printing metrics for each fold (River model)
for i in range(len(mse_list_river)):
    print(f"River SNARIMAX Model - Fold {i+1} - MSE: {mse_list_river[i]}, RMSE: {rmse_list_river[i]}, MAE: {mae_list_river[i]}")

# Calculating and printing mean and standard deviation of metrics (River model)
print("\nRiver SNARIMAX Model - Average Metrics and Standard Deviations:")
print(f"MSE: Mean = {np.mean(mse_list_river)}, Std = {np.std(mse_list_river)}")
print(f"RMSE: Mean = {np.mean(rmse_list_river)}, Std = {np.std(rmse_list_river)}")
print(f"MAE: Mean = {np.mean(mae_list_river)}, Std = {np.std(mae_list_river)}")
```

```python
mse_list_sarima, rmse_list_sarima, mae_list_sarima = [], [], []
all_y_test_sarima, all_y_pred_sarima = [], []

# Iterate over folds for SARIMA model
for train_index, test_index in tscv.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit SARIMA model
    sarima_model = SARIMAX(y_train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 365))
    sarima_result = sarima_model.fit(disp=False)

    # Predictions
    y_pred = sarima_result.forecast(steps=len(y_test))

    # Metrics calculation
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)

    mse_list_sarima.append(mse)
    rmse_list_sarima.append(rmse)
    mae_list_sarima.append(mae)

    all_y_test_sarima.extend(y_test)
    all_y_pred_sarima.extend(y_pred)

# Printing metrics for each fold (SARIMA model)
for i in range(len(mse_list_sarima)):
    print(f"SARIMA Model - Fold {i+1} - MSE: {mse_list_sarima[i]}, RMSE: {rmse_list_sarima[i]}, MAE: {mae_list_sarima[i]}")

# Calculating and printing mean and standard deviation of metrics (SARIMA model)
print("\nSARIMA Model - Average Metrics and Standard Deviations:")
print(f"MSE: Mean = {np.mean(mse_list_sarima)}, Std = {np.std(mse_list_sarima)}")
print(f"RMSE: Mean = {np.mean(rmse_list_sarima)}, Std = {np.std(rmse_list_sarima)}")
print(f"MAE: Mean = {np.mean(mae_list_sarima)}, Std = {np.std(mae_list_sarima)}")
```
